```python
df = pd.read_csv("data_to_train.csv", index_col=0)
df
```

```python
y_df = df["price"]
x_df = df.drop(["price"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_df)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

X, y = x_scaled, y_df.to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#Linear Regression
dtr = DecisionTreeRegressor()



dtr.fit(X_train, y_train)
y_pred = dtr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error of Linear Regression on test set: {mae:.2f}')
print(f'Mean Squared Error of Linear Regression on test set: {mse:.2f}')
print(f'R-squared Error of Linear Regression on test set: {r2:.6f}')
print(f'Root Mean Squared Error of Linear Regression on test set: {rmse:.2f}')
```

1. Mean Absolute Error (MAE): 0.5332

    An MAE of 0.5332 means that, on average, the models predictions are approximately $0.5332 away from the true house prices.

2. Mean Squared Error (MSE): 0.5559

    An MSE of 0.5559 means that, on average, the squared prediction errors are approximately 0.5559.

3. R-squared (R): 0.5758

    An R of 0.5758 indicates that the model can explain approximately 57.58% of the variance in house prices.

4. Root Mean Squared Error (RMSE): 0.7456

    An RMSE of 0.7456 indicates that, on average, the models predictions have an error of approximately $0.7456 in the same units as the house prices.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#Linear Regression
lr = LinearRegression()



lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error of Linear Regression on test set: {mae:.2f}')
print(f'Mean Squared Error of Linear Regression on test set: {mse:.2f}')
print(f'R-squared Error of Linear Regression on test set: {r2:.6f}')
print(f'Root Mean Squared Error of Linear Regression on test set: {rmse:.2f}')
```
